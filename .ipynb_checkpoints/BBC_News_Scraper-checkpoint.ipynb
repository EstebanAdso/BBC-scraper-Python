# Scraping de Noticias de Tecnología BBC

Este notebook realiza el scraping de artículos de la sección de tecnología de la BBC, almacenando los resultados tanto en una base de datos SQLite como en archivos CSV. El objetivo principal es mantener un registro histórico de noticias tecnológicas, evitando la inserción de duplicados y asegurando la organización de los archivos generados.

## Estructura del proyecto
- Extracción de datos de la web de BBC Technology
- Transformación y limpieza de datos
- Validación de artículos
- Almacenamiento en base de datos SQLite y CSV
- Análisis y visualización de los datos obtenidos

## 1. Configuración inicial

Importamos las bibliotecas necesarias y configuramos el entorno de trabajo.

```python
# Importación de bibliotecas necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
import os
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from urllib.parse import urljoin
from typing import List, Dict, Optional
import re

# Configuración de visualizaciones
plt.style.use('ggplot')
sns.set(style="whitegrid")

# Configuración de directorios
output_dir = os.path.join(os.path.dirname(os.getcwd()), 'ArchivosGenerados')
os.makedirs(output_dir, exist_ok=True)
print(f"Directorio de salida: {output_dir}")
```

## 2. Definición de clases y modelos

Definimos la clase Article para representar cada artículo de noticias.

```python
from dataclasses import dataclass

@dataclass
class Article:
    title: str
    summary: str
    link: str
    date: str
    source: str = 'BBC Technology'
    
    @property
    def to_dict(self):
        return {
            'title': self.title,
            'summary': self.summary,
            'link': self.link,
            'date': self.date,
            'source': self.source
        }
```

## 3. Implementación del Scraper

Creamos la clase BBCScraper para extraer, transformar y validar los datos de la web de BBC.

```python
class BBCScraper:
    def __init__(self):
        self.BASE_URL = "https://www.bbc.com"
        self.TECH_URL = f"{self.BASE_URL}/news/technology"
        self.HEADERS = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
    
    def extract(self) -> Optional[str]:
        print("Extracting data from BBC Technology...")
        try:
            response = requests.get(self.TECH_URL, headers=self.HEADERS, timeout=15)
            response.raise_for_status()
            return response.text
        except Exception as e:
            print(f"Extraction error: {e}")
            return None
    
    def transform(self, html_content: str) -> List[Article]:
        print("Transforming data...")
        soup = BeautifulSoup(html_content, 'html.parser')
        articles = []
        
        news_containers = soup.find_all('div', {'data-testid': 'edinburgh-card'}) or \
                         soup.find_all('div', class_='ssrcss-1f3bvyz-Stack')
        
        for container in news_containers:
            try:
                title = container.find('h2').get_text(strip=True) if container.find('h2') else "No title"
                
                link = "#"
                link_tag = container.find('a', href=True)
                if link_tag:
                    link = urljoin(self.BASE_URL, link_tag['href'])
                
                summary = container.find('p').get_text(strip=True) if container.find('p') else "No summary"
                
                articles.append(Article(
                    title=title,
                    summary=summary,
                    link=link,
                    date=datetime.now().strftime('%Y-%m-%d')
                ))
            except Exception as e:
                print(f"Error processing article: {e}")
                continue
        
        return articles
    
    def validate(self, articles: List[Article]) -> List[Article]:
        print("Validating data...")
        valid_articles = []
        
        for article in articles:
            if article.title and article.title != "No title" and article.date:
                valid_articles.append(article)
            else:
                print(f"Invalid article discarded: {article.title}")
        
        print(f"Valid articles: {len(valid_articles)}/{len(articles)}")
        return valid_articles
```

## 4. Extracción de datos

Ejecutamos el scraper para extraer los datos de la web de BBC Technology.

```python
# Inicializar y ejecutar el scraper
scraper = BBCScraper()
html_content = scraper.extract()

if html_content:
    print("HTML extraído correctamente")
    # Mostrar las primeras 200 caracteres del HTML para verificar
    print(html_content[:200] + "...")
else:
    print("Error al extraer HTML")
```

## 5. Transformación y validación de datos

Procesamos el HTML extraído para obtener los artículos y validamos su contenido.

```python
# Transformar el HTML en objetos Article
if html_content:
    raw_articles = scraper.transform(html_content)
    print(f"Se encontraron {len(raw_articles)} artículos en total")
    
    # Mostrar los primeros 3 artículos para verificar
    for i, article in enumerate(raw_articles[:3]):
        print(f"\nArtículo {i+1}:")
        print(f"Título: {article.title}")
        print(f"Resumen: {article.summary}")
        print(f"Link: {article.link}")
        print(f"Fecha: {article.date}")
    
    # Validar los artículos
    valid_articles = scraper.validate(raw_articles)
    print(f"\nArtículos válidos: {len(valid_articles)}")
else:
    valid_articles = []
```

## 6. Almacenamiento de datos

Implementamos funciones para guardar los artículos en CSV y en una base de datos SQLite.

```python
# Función para guardar en CSV
def save_to_csv(articles: List[Article]) -> str:
    print("Creando archivo CSV...")
    timestamp = datetime.now().strftime('%Y%m%d_%H%M')
    filename = f"bbc_tech_{timestamp}.csv"
    filepath = os.path.join(output_dir, filename)
    
    if not articles:
        print("No hay artículos para guardar")
        return ""
    
    # Convertir a DataFrame para mejor manipulación
    df = pd.DataFrame([article.to_dict for article in articles])
    
    # Guardar a CSV
    df.to_csv(filepath, index=False, encoding='utf-8')
    print(f"Datos guardados en {filepath}")
    return filepath

# Función para guardar en SQLite
def save_to_sqlite(articles: List[Article], db_name: str = "bbc_sqlite.db"):
    print("Guardando en base de datos SQLite...")
    db_path = os.path.join(output_dir, db_name)
    
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Crear tabla si no existe
        create_table_query = """
        CREATE TABLE IF NOT EXISTS tech_articles (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            title TEXT NOT NULL,
            summary TEXT,
            link TEXT UNIQUE,
            date TEXT NOT NULL,
            source TEXT,
            timestamp DATETIME DEFAULT CURRENT_TIMESTAMP
        );
        """
        cursor.execute(create_table_query)
        cursor.execute("CREATE UNIQUE INDEX IF NOT EXISTS idx_unique_link ON tech_articles(link)")
        
        # Insertar artículos
        insert_query = """
        INSERT OR IGNORE INTO tech_articles (title, summary, link, date, source)
        VALUES (?, ?, ?, ?, ?)
        """
        nuevos = 0
        for article in articles:
            cursor.execute(insert_query, (
                article.title,
                article.summary,
                article.link,
                article.date,
                article.source
            ))
            if cursor.rowcount > 0:
                nuevos += 1
        
        conn.commit()
        print(f"Datos guardados en SQLite ({db_path})")
        print(f"Total de registros insertados: {nuevos}")
        
        return db_path
    except Exception as e:
        print(f"Error al guardar en SQLite: {e}")
        return None
    finally:
        if 'conn' in locals() and conn:
            conn.close()

# Guardar los artículos si existen
if 'valid_articles' in locals() and valid_articles:
    csv_path = save_to_csv(valid_articles)
    db_path = save_to_sqlite(valid_articles)
else:
    print("No hay artículos válidos para guardar")
```

## 7. Carga de datos para análisis

Cargamos los datos desde la base de datos SQLite para su análisis.

```python
# Función para cargar datos desde SQLite
def load_from_sqlite(db_path):
    try:
        conn = sqlite3.connect(db_path)
        query = "SELECT * FROM tech_articles"
        df = pd.read_sql_query(query, conn)
        conn.close()
        return df
    except Exception as e:
        print(f"Error al cargar datos desde SQLite: {e}")
        return None

# Cargar datos desde la base de datos o usar los artículos recién extraídos
if 'db_path' in locals() and os.path.exists(db_path):
    df_articles = load_from_sqlite(db_path)
    print(f"Se cargaron {len(df_articles)} artículos desde la base de datos")
elif 'valid_articles' in locals() and valid_articles:
    # Si no hay base de datos, usar los artículos recién extraídos
    df_articles = pd.DataFrame([article.to_dict for article in valid_articles])
    print("Usando artículos recién extraídos para el análisis")
else:
    print("No hay datos disponibles para análisis")
    df_articles = pd.DataFrame()

# Mostrar información básica del DataFrame
if not df_articles.empty:
    print("\nInformación del DataFrame:")
    print(df_articles.info())
    
    print("\nPrimeras 5 filas:")
    display(df_articles.head())
```

## 8. Análisis exploratorio de datos

Realizamos un análisis básico de los datos recopilados.

```python
# Análisis básico de los datos
if not df_articles.empty:
    # Estadísticas descriptivas
    print("Estadísticas de longitud de títulos:")
    title_lengths = df_articles['title'].str.len()
    print(title_lengths.describe())
    
    print("\nEstadísticas de longitud de resúmenes:")
    summary_lengths = df_articles['summary'].str.len()
    print(summary_lengths.describe())
    
    # Conteo de artículos por fecha
    if 'date' in df_articles.columns:
        print("\nNúmero de artículos por fecha:")
        date_counts = df_articles['date'].value_counts().sort_index()
        print(date_counts)
```

## 9. Visualizaciones

Creamos visualizaciones para entender mejor los datos.

```python
# Visualizaciones
if not df_articles.empty:
    # 1. Distribución de artículos por fecha
    if 'date' in df_articles.columns and len(df_articles['date'].unique()) > 1:
        plt.figure(figsize=(12, 6))
        date_counts = df_articles['date'].value_counts().sort_index()
        sns.barplot(x=date_counts.index, y=date_counts.values)
        plt.title('Número de artículos por fecha')
        plt.xlabel('Fecha')
        plt.ylabel('Cantidad de artículos')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    # 2. Longitud de títulos y resúmenes
    plt.figure(figsize=(15, 6))
    
    plt.subplot(1, 2, 1)
    title_lengths = df_articles['title'].str.len()
    sns.histplot(title_lengths, kde=True)
    plt.title('Distribución de longitud de títulos')
    plt.xlabel('Número de caracteres')
    plt.ylabel('Frecuencia')
    
    plt.subplot(1, 2, 2)
    summary_lengths = df_articles['summary'].str.len()
    sns.histplot(summary_lengths, kde=True)
    plt.title('Distribución de longitud de resúmenes')
    plt.xlabel('Número de caracteres')
    plt.ylabel('Frecuencia')
    
    plt.tight_layout()
    plt.show()
```

## 10. Análisis de texto

Realizamos un análisis de texto básico para identificar palabras frecuentes en títulos y resúmenes.

```python
# Instalamos wordcloud si no está disponible
try:
    import wordcloud
except ImportError:
    !pip install wordcloud
    import wordcloud

# Instalamos nltk si no está disponible
try:
    import nltk
    from nltk.corpus import stopwords
    nltk.download('stopwords', quiet=True)
except ImportError:
    !pip install nltk
    import nltk
    nltk.download('stopwords', quiet=True)
    from nltk.corpus import stopwords

from wordcloud import WordCloud

# Función para generar nube de palabras
def generate_wordcloud(text, title):
    stop_words = set(stopwords.words('english'))
    
    # Preprocesamiento del texto
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Eliminar puntuación
    words = text.split()
    words = [word for word in words if word not in stop_words]
    text = ' '.join(words)
    
    # Generar nube de palabras
    wordcloud = WordCloud(width=800, height=400, background_color='white', 
                          max_words=100, contour_width=3, contour_color='steelblue')
    wordcloud.generate(text)
    
    # Mostrar
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Generar nubes de palabras si hay datos disponibles
if not df_articles.empty:
    # Combinar todos los títulos y resúmenes
    all_titles = ' '.join(df_articles['title'].tolist())
    all_summaries = ' '.join(df_articles['summary'].tolist())
    
    # Generar nubes de palabras
    generate_wordcloud(all_titles, 'Palabras más frecuentes en títulos')
    generate_wordcloud(all_summaries, 'Palabras más frecuentes en resúmenes')
```

## 11. Análisis de sentimiento (opcional)

Realizamos un análisis de sentimiento básico de los títulos y resúmenes.

```python
# Análisis de sentimiento básico usando TextBlob
try:
    from textblob import TextBlob
except ImportError:
    !pip install textblob
    from textblob import TextBlob

# Descargar recursos necesarios para TextBlob
import subprocess
try:
    subprocess.run(["python", "-m", "textblob.download_corpora"], check=True, capture_output=True)
except:
    print("No se pudieron descargar los recursos de TextBlob, pero intentaremos continuar.")

if not df_articles.empty:
    # Función para obtener polaridad (positivo/negativo) y subjetividad
    def get_sentiment(text):
        blob = TextBlob(text)
        return blob.sentiment.polarity, blob.sentiment.subjectivity
    
    # Aplicar a títulos y resúmenes
    df_articles['title_polarity'], df_articles['title_subjectivity'] = zip(*df_articles['title'].apply(get_sentiment))
    df_articles['summary_polarity'], df_articles['summary_subjectivity'] = zip(*df_articles['summary'].apply(get_sentiment))
    
    # Visualizar distribución de polaridad
    plt.figure(figsize=(15, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(df_articles['title_polarity'], kde=True)
    plt.title('Polaridad de títulos')
    plt.xlabel('Polaridad (-1: negativo, 1: positivo)')
    plt.ylabel('Frecuencia')
    
    plt.subplot(1, 2, 2)
    sns.histplot(df_articles['summary_polarity'], kde=True)
    plt.title('Polaridad de resúmenes')
    plt.xlabel('Polaridad (-1: negativo, 1: positivo)')
    plt.ylabel('Frecuencia')
    
    plt.tight_layout()
    plt.show()
    
    # Gráfico de dispersión de polaridad vs. subjetividad
    plt.figure(figsize=(10, 6))
    plt.scatter(df_articles['summary_polarity'], df_articles['summary_subjectivity'], alpha=0.7)
    plt.title('Polaridad vs. Subjetividad en resúmenes')
    plt.xlabel('Polaridad (-1: negativo, 1: positivo)')
    plt.ylabel('Subjetividad (0: objetivo, 1: subjetivo)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## 12. Conclusiones

Resumen de los hallazgos del análisis.

```python
# Resumen de hallazgos
if not df_articles.empty:
    print("Resumen del análisis de artículos de tecnología de BBC:")
    print(f"- Total de artículos analizados: {len(df_articles)}")
    
    if 'date' in df_articles.columns:
        print(f"- Rango de fechas: {df_articles['date'].min()} a {df_articles['date'].max()}")
    
    print(f"- Longitud promedio de títulos: {df_articles['title'].str.len().mean():.1f} caracteres")
    print(f"- Longitud promedio de resúmenes: {df_articles['summary'].str.len().mean():.1f} caracteres")

    if 'title_polarity' in df_articles.columns:
        print(f"- Polaridad promedio de títulos: {df_articles['title_polarity'].mean():.3f}")
        print(f"- Polaridad promedio de resúmenes: {df_articles['summary_polarity'].mean():.3f}")
        
        # Clasificar artículos por sentimiento
        sentiment_categories = ['Negativo', 'Neutral', 'Positivo']
        df_articles['sentiment_category'] = pd.cut(
            df_articles['summary_polarity'], 
            bins=[-1.1, -0.1, 0.1, 1.1], 
            labels=sentiment_categories
        )
        
        sentiment_counts = df_articles['sentiment_category'].value_counts()
        print("\nDistribución de sentimiento en artículos:")
        for category in sentiment_categories:
            if category in sentiment_counts:
                print(f"- {category}: {sentiment_counts[category]} artículos ({sentiment_counts[category]/len(df_articles)*100:.1f}%)")

    # Guardar el DataFrame analizado
    analysis_file = os.path.join(output_dir, "bbc_tech_analysis.csv")
    df_articles.to_csv(analysis_file, index=False)
    print(f"\nAnálisis guardado en: {analysis_file}")
```
